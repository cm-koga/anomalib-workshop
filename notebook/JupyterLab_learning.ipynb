# JupyterLabの概要
JupyterLabはWebブラウザ上で動作するインタラクティブな開発環境です。
* Webブラウザ上に表示されたノートにはセルがあります。セルにPythonコードを記述し、実行結果を表示することができます。
* Markdownでドキュメントを書くことができます（画面上のプルダウンでセルをMarkdown用にするかコード用にするかを選択することができます）

### セルの最後にある式の評価結果が表示される
JupyterLabでは、コードの最後に評価した式の結果が表示されます（下記セルだと 最終行にある a @ bの結果が表示されます）

下のセルを実行してみてください。

````python
# ２つの行列の行列積を計算
import numpy as np

a = np.array([[3.1, 2.5], [1.2, 3.1], [5.2, 2.5]])
b = np.array([[1.2, 2.1, 4.8], [9.2, 3.4, 6.2]])

a @ b
````

## セルをコマンドプロンプトのようにつかう
コードセルで頭に!マークをつけて、コマンドプロンプトのコマンドを実行することができます。
たとえばdirコマンドをセルで実行したい場合は
```
!dir
```
とします。

試しに下のセルで実行してみましょう。dirコマンドの結果が表示されるはずです。

````python
!dir
````

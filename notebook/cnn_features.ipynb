# CNN特徴量

```python
import torch
import torchvision
from torchvision import transforms
from torchvision.models.feature_extraction import create_feature_extractor
import cv2

# ResNet18 Imagenet学習済モデル
model = torchvision.models.resnet18(weights="IMAGENET1K_V1")

# 画像読み込み
im = cv2.imread("datasets/wood/test/NG/IMG_3792_0101.png")
print("image shape", im.shape)

# 前処理
transform_list = [
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    transforms.Resize((224, 224)),
]
transform = transforms.Compose(transform_list)

x = transform(im).unsqueeze(0)

print("preprocess shape", x.shape)

# 推論
with torch.no_grad():
    outputs = model(x)

print("resnet18 output shape", outputs.shape)
```

# 中間特徴を抽出

```python
import torch
import torchvision
from torchvision import transforms
from torchvision.models.feature_extraction import create_feature_extractor
import cv2

# ResNet18 Imagenet学習済モデル
model = torchvision.models.resnet18(weights="IMAGENET1K_V1")
layers = ['layer2', 'layer3']
extractor = create_feature_extractor(model, layers) # layer2とlayer3の出力を最終出力とする

# 画像読み込み
im = cv2.imread("datasets/wood/test/NG/IMG_3792_0101.png")
print("image shape", im.shape)

# 前処理
transform_list = [
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    transforms.Resize((224, 224)),
]
transform = transforms.Compose(transform_list)

x = transform(im).unsqueeze(0)

print("preprocess shape", x.shape)

# 中間特徴を抽出
with torch.no_grad():
    outputs =extractor(x)

print("--- outputs ---")
print("output dict keys", outputs.keys())
for k, v in outputs.items():
    print(f"{k}: {v.shape}")
```
